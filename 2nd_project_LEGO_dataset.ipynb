```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv  ')
themes = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')

themes.head()
```

####Task 1: what percentage of all licensed set ever relesed were star wars themed?

```python
themes[themes['is_licensed'] == True];
```

```python
merged = df.merge(themes, left_on= 'parent_theme', right_on= 'name')

merged.drop('name_y', axis=1, inplace= True)

merged.rename(columns ={'name_x': 'name'}, inplace = True)

merged.head(50)
```

```python
licensed = merged[merged['is_licensed'] == True]

licensed.dropna(subset = ['set_num'] , inplace = True)

star_wars = licensed[licensed ['parent_theme'] == 'Star Wars']



star_wars.shape[0]/ licensed.shape[0]
```

####Task2 : In which year was star wars not the most populer licensed theme(in term of number of sets released that year)?




```python
licensed['count'] = 1
licensed = licensed.sort_values('year')
# licensed= licensed.
max = licensed.groupby(['year', 'parent_theme']).sum().reset_index()
max_df = max.sort_values('count', ascending= False). drop_duplicates(['year'])
max_df.sort_values('year' , inplace = True)
max_df
# max['']
# max.drop_duplicates(['year'])
```
